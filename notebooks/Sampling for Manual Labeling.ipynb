```python
import numpy as np
import pandas as pd

# Tambah kolom panjang teks (jumlah kata)
df["word_count"] = df["content"].str.split().str.len()

# Kategorikan panjang ulasan
df["length_cat"] = pd.cut(
    df["word_count"],
    bins=[0, 10, 30, 10000],
    labels=["pendek", "sedang", "panjang"]
)

# Deteksi keyword promosi
promo_kw = r"(promo|diskon|gratis|kode|gunakan|voucher|hemat|potongan|cashback)"
df["has_promo"] = df["content"].str.lower().str.contains(promo_kw, regex=True)

# Tambahkan kategori rating agar seimbang
df["score_cat"] = df["score"].clip(1, 5).astype(str)

# Gabungkan faktor-faktor untuk stratifikasi
group_cols = ["score_cat", "length_cat", "has_promo"]

# Sampling proporsional dari setiap kombinasi
target_n = 600
samples = []

for _, g in df.groupby(group_cols, dropna=False):
    if len(g) == 0:
        continue
    k = max(1, int(target_n * len(g) / len(df)))
    samples.append(g.sample(n=min(k, len(g)), random_state=42))

# Gabungkan hasil sampling dari semua grup
sampled_df = pd.concat(samples)

# Jika hasil kurang dari target, ambil tambahan dari sisa data tanpa duplikat
if len(sampled_df) < target_n:
    remaining = df.drop(sampled_df.index)
    extra = remaining.sample(n=target_n - len(sampled_df), random_state=42)
    sampled_df = pd.concat([sampled_df, extra])

# Jika hasil lebih dari target, lakukan random trimming tanpa replacement
elif len(sampled_df) > target_n:
    sampled_df = sampled_df.sample(n=target_n, random_state=42, replace=False)

# Cek jumlah duplikat
dup_count = sampled_df.duplicated().sum()
print(f"Jumlah duplikat: {dup_count}")
```
