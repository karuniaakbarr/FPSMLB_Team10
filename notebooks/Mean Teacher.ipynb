# Mean Teacher

```python
# ===========================================
# 0. IMPORT LIBRARY
# ===========================================
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

import pandas as pd
import numpy as np

from transformers import AutoTokenizer, AutoModel
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import classification_report, roc_auc_score

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)
```

```python
# ===========================================
# 1. PERSIAPAN DATAFRAME & TIPE DATA
# ===========================================
# Asumsi: df sudah ada di memory dengan kolom:
# content, score_cat, thumbsUpCount, word_count, has_promo, label

df = df.copy()

# Pastikan tipe data sesuai kebutuhan
df["content"] = df["content"].astype(str)
df["score_cat"] = df["score_cat"].astype(str)        # untuk OneHotEncoder
df["thumbsUpCount"] = df["thumbsUpCount"].astype(float)
df["word_count"] = df["word_count"].astype(float)
df["has_promo"] = df["has_promo"].astype(int)        # bool -> int 0/1

# label: category -> Int64 (boleh 0,1,NaN)
df["label"] = df["label"].astype("Int64")

print(df.dtypes)
print(df["label"].value_counts(dropna=False))
```

```python
# ===========================================
# 2. PISAH LABELED & UNLABELED + TRAIN/VAL/TEST
# ===========================================

# labeled: label tidak NaN
labeled_df = df[~df["label"].isna()].reset_index(drop=True)
unlabeled_df = df[df["label"].isna()].reset_index(drop=True)

print("Labeled:", len(labeled_df), " | Unlabeled:", len(unlabeled_df))

# stratified split labeled -> train / test
train_labeled, test_labeled = train_test_split(
    labeled_df,
    test_size=0.2,
    stratify=labeled_df["label"],
    random_state=42
)

# train -> train / val
train_labeled, val_labeled = train_test_split(
    train_labeled,
    test_size=0.1,
    stratify=train_labeled["label"],
    random_state=42
)

print("Train labeled:", len(train_labeled))
print("Val labeled  :", len(val_labeled))
print("Test labeled :", len(test_labeled))
```

```python
# ===========================================
# 3. ENCODING METADATA (score_cat, thumbsUpCount, word_count, has_promo)
# ===========================================
meta_cols_cat = ["score_cat"]
meta_cols_num = ["thumbsUpCount", "word_count", "has_promo"]

ohe = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
scaler = StandardScaler()

# Fit encoder dan scaler di TRAIN saja (untuk hindari data leakage)
X_cat_train = ohe.fit_transform(train_labeled[meta_cols_cat])
X_num_train = scaler.fit_transform(train_labeled[meta_cols_num])

meta_dim = X_cat_train.shape[1] + X_num_train.shape[1]
print("Dimensi fitur metadata:", meta_dim)

def make_meta_matrix(df_part):
    X_cat = ohe.transform(df_part[meta_cols_cat])
    X_num = scaler.transform(df_part[meta_cols_num])
    return np.concatenate([X_cat, X_num], axis=1).astype(np.float32)

# Buat matrix meta untuk semua subset
train_meta = make_meta_matrix(train_labeled)
val_meta   = make_meta_matrix(val_labeled)
test_meta  = make_meta_matrix(test_labeled)

unlabeled_meta = make_meta_matrix(unlabeled_df) if len(unlabeled_df) > 0 else None
```

```python
# ===========================================
# 4. TOKENIZER & INDOBERTWEET BASE
# ===========================================
# IndoBERTweet (bahasa informal / medsos)

MODEL_NAME = "indolem/indobertweet-base-uncased"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
bert_base = AutoModel.from_pretrained(MODEL_NAME)
bert_base = bert_base.to(device)

print("Hidden size:", bert_base.config.hidden_size)
```

```python
# ===========================================
# 5. DATASET CLASS + DATALOADER
# ===========================================
MAX_LEN = 64  # bisa kamu sesuaikan

class ReviewDataset(Dataset):
    def __init__(self, df_part, meta_matrix, tokenizer, max_len, has_labels=True):
        self.texts = df_part["content"].tolist()
        self.meta = meta_matrix
        self.tokenizer = tokenizer
        self.max_len = max_len
        self.has_labels = has_labels
        if has_labels:
            # label Int64 -> int np.array
            self.labels = df_part["label"].astype(int).values
        else:
            self.labels = None

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])
        meta = torch.tensor(self.meta[idx], dtype=torch.float32)

        enc = self.tokenizer(
            text,
            padding="max_length",
            truncation=True,
            max_length=self.max_len,
            return_tensors="pt"
        )

        item = {
            "input_ids": enc["input_ids"].squeeze(0),
            "attention_mask": enc["attention_mask"].squeeze(0),
            "meta": meta
        }

        if self.has_labels:
            item["labels"] = torch.tensor(self.labels[idx], dtype=torch.long)

        return item

# Buat dataset
train_labeled_ds = ReviewDataset(train_labeled, train_meta, tokenizer, MAX_LEN, has_labels=True)
val_ds           = ReviewDataset(val_labeled,   val_meta,   tokenizer, MAX_LEN, has_labels=True)
test_ds          = ReviewDataset(test_labeled,  test_meta,  tokenizer, MAX_LEN, has_labels=True)

if len(unlabeled_df) > 0:
    unlabeled_ds = ReviewDataset(unlabeled_df, unlabeled_meta, tokenizer, MAX_LEN, has_labels=False)
else:
    unlabeled_ds = None

# DataLoader
BATCH_LABELED = 8
BATCH_UNLABELED = 16

train_labeled_loader = DataLoader(train_labeled_ds, batch_size=BATCH_LABELED, shuffle=True)
val_loader           = DataLoader(val_ds,          batch_size=64, shuffle=False)
test_loader          = DataLoader(test_ds,         batch_size=64, shuffle=False)

if unlabeled_ds is not None:
    # Ambil hanya 500 data unlabeled per epoch
    unlabeled_small = unlabeled_df.sample(n=500, random_state=epoch, replace=False)
    unlabeled_meta_small = make_meta_matrix(unlabeled_small)

    unlabeled_ds = ReviewDataset(unlabeled_small, unlabeled_meta_small, tokenizer, MAX_LEN, has_labels=False)
    unlabeled_loader = DataLoader(unlabeled_ds, batch_size=BATCH_UNLABELED, shuffle=True)
else:
    unlabeled_loader = None
```

```python
# ===========================================
# 6. MODEL FUSION (BERT + METADATA)
# ===========================================
class FusionClassifier(nn.Module):
    def __init__(self, bert_model, meta_dim, hidden_size=256, num_labels=2):
        super().__init__()
        self.bert = bert_model
        bert_hidden = bert_model.config.hidden_size

        self.classifier = nn.Sequential(
            nn.Linear(bert_hidden + meta_dim, hidden_size),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(hidden_size, num_labels)
        )

    def forward(self, input_ids, attention_mask, meta):
        outputs = self.bert(
            input_ids=input_ids,
            attention_mask=attention_mask
        )
        # CLS token embedding
        cls_emb = outputs.last_hidden_state[:, 0, :]  # [batch, hidden]

        x = torch.cat([cls_emb, meta], dim=1)
        logits = self.classifier(x)
        return logits

student = FusionClassifier(bert_base, meta_dim=meta_dim).to(device)

import copy
teacher = copy.deepcopy(student).to(device)

# teacher tidak di-backprop
for p in teacher.parameters():
    p.requires_grad = False

def update_teacher(student, teacher, alpha=0.99):
    for t_param, s_param in zip(teacher.parameters(), student.parameters()):
        t_param.data.mul_(alpha).add_(s_param.data, alpha=1 - alpha)
```

```python
# ===========================================
# 7. TRAINING SETUP (LOSS, OPTIMIZER, HYPERPARAMETER)
# ===========================================
from torch.optim import AdamW

ce_loss = nn.CrossEntropyLoss()
optimizer = AdamW(student.parameters(), lr=2e-5)

NUM_EPOCHS = 5
LAMBDA_U = 10.0     # bobot consistency loss
ALPHA_EMA = 0.99    # koefisien EMA teacher
```

```python
# ===========================================
# 8. TRAINING LOOP MEAN TEACHER
# ===========================================
for epoch in range(NUM_EPOCHS):
    student.train()
    teacher.train()

    total_sup = 0.0
    total_cons = 0.0

    if unlabeled_loader is not None:
        unlabeled_iter = iter(unlabeled_loader)
    else:
        unlabeled_iter = None

    for batch in train_labeled_loader:
        # ----- Labeled batch -----
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        meta = batch["meta"].to(device)
        labels = batch["labels"].to(device)

        logits_s = student(input_ids, attention_mask, meta)
        sup_loss = ce_loss(logits_s, labels)

        # ----- Unlabeled batch: consistency loss -----
        cons_loss = torch.tensor(0.0, device=device)

        if unlabeled_iter is not None:
            try:
                ubatch = next(unlabeled_iter)
            except StopIteration:
                unlabeled_iter = iter(unlabeled_loader)
                ubatch = next(unlabeled_iter)

            u_input_ids = ubatch["input_ids"].to(device)
            u_attention_mask = ubatch["attention_mask"].to(device)
            u_meta = ubatch["meta"].to(device)

            with torch.no_grad():
                teacher_logits = teacher(u_input_ids, u_attention_mask, u_meta)
                teacher_probs = F.softmax(teacher_logits, dim=1)

            student_logits_u = student(u_input_ids, u_attention_mask, u_meta)
            student_probs = F.softmax(student_logits_u, dim=1)

            cons_loss = F.mse_loss(student_probs, teacher_probs)

        loss = sup_loss + LAMBDA_U * cons_loss

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # update teacher dengan EMA
        if unlabeled_iter is not None:
            update_teacher(student, teacher, alpha=ALPHA_EMA)

        total_sup += sup_loss.item()
        total_cons += cons_loss.item()

    avg_sup = total_sup / len(train_labeled_loader)
    avg_cons = total_cons / len(train_labeled_loader)
    print(f"Epoch {epoch+1}/{NUM_EPOCHS} | Supervised loss: {avg_sup:.4f} | Consistency loss: {avg_cons:.4f}")
```

```python
# ===========================================
# 9. FUNGSI EVALUASI
# ===========================================
def evaluate(model, loader, name="MODEL"):
    model.eval()
    all_preds = []
    all_probs = []
    all_labels = []

    with torch.no_grad():
        for batch in loader:
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            meta = batch["meta"].to(device)
            labels = batch["labels"].to(device)

            logits = model(input_ids, attention_mask, meta)
            probs = F.softmax(logits, dim=1)[:, 1]
            preds = torch.argmax(logits, dim=1)

            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())

    print("=== Evaluation:", name, "===")
    print(classification_report(all_labels, all_preds, digits=4))
    try:
        print("ROC-AUC:", roc_auc_score(all_labels, all_probs))
    except:
        print("ROC-AUC tidak bisa dihitung (label hanya 1 kelas).")
```

```python
# ===========================================
# 10. EVALUASI STUDENT DAN TEACHER
# ===========================================
evaluate(student, test_loader, name="STUDENT")
evaluate(teacher, test_loader, name="TEACHER")
```
