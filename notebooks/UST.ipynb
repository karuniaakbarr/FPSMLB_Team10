# Uncertainty aware Self-Training

## Import Library

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

import pandas as pd
import numpy as np

from transformers import AutoTokenizer, AutoModel
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import classification_report, roc_auc_score

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)
```

## Rapikan tipe data di df

```python
df = df.copy()

# Pastikan tipe data sesuai
df["content"] = df["content"].astype(str)
df["score_cat"] = df["score_cat"].astype(str)      # untuk OneHotEncoder
df["thumbsUpCount"] = df["thumbsUpCount"].astype(float)
df["word_count"] = df["word_count"].astype(float)
df["has_promo"] = df["has_promo"].astype(int)      # bool -> 0/1

# label boleh 0,1,NaN
df["label"] = df["label"].astype("Int64")

print(df.dtypes)
print(df["label"].value_counts(dropna=False))
```

## Split labeled / unlabeled + train / val / test

```python
# Data berlabel dan tidak berlabel
labeled_df = df[~df["label"].isna()].reset_index(drop=True)
unlabeled_df = df[df["label"].isna()].reset_index(drop=True)

print("Labeled:", len(labeled_df), "| Unlabeled:", len(unlabeled_df))

# Split labeled → train / test
train_labeled, test_labeled = train_test_split(
    labeled_df,
    test_size=0.2,
    stratify=labeled_df["label"],
    random_state=42
)

# Split train → train / val
train_labeled, val_labeled = train_test_split(
    train_labeled,
    test_size=0.1,
    stratify=train_labeled["label"],
    random_state=42
)

print("Train labeled:", len(train_labeled))
print("Val labeled  :", len(val_labeled))
print("Test labeled :", len(test_labeled))
```

## Encoding metadata

```python
meta_cols_cat = ["score_cat"]
meta_cols_num = ["thumbsUpCount", "word_count", "has_promo"]

ohe = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
scaler = StandardScaler()

# Fit encoder & scaler di train_labeled saja
X_cat_train = ohe.fit_transform(train_labeled[meta_cols_cat])
X_num_train = scaler.fit_transform(train_labeled[meta_cols_num])

meta_dim = X_cat_train.shape[1] + X_num_train.shape[1]
print("Dimensi fitur metadata:", meta_dim)

def make_meta_matrix(df_part):
    X_cat = ohe.transform(df_part[meta_cols_cat])
    X_num = scaler.transform(df_part[meta_cols_num])
    return np.concatenate([X_cat, X_num], axis=1).astype(np.float32)

# Metadata untuk train / val / test
train_meta = make_meta_matrix(train_labeled)
val_meta   = make_meta_matrix(val_labeled)
test_meta  = make_meta_matrix(test_labeled)
```

## Load IndoBERTweet

```python
MODEL_NAME = "indolem/indobertweet-base-uncased"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
bert_base = AutoModel.from_pretrained(MODEL_NAME).to(device)

print("Hidden size:", bert_base.config.hidden_size)
```

## Dataset dan DataLoader

```python
MAX_LEN = 64  # dipendekkan biar training lebih cepat

class ReviewDataset(Dataset):
    def __init__(self, df_part, meta_matrix, tokenizer, max_len, has_labels=True):
        self.texts = df_part["content"].tolist()
        self.meta = meta_matrix
        self.tokenizer = tokenizer
        self.max_len = max_len
        self.has_labels = has_labels
        if has_labels:
            self.labels = df_part["label"].astype(int).values
        else:
            self.labels = None

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        meta = torch.tensor(self.meta[idx], dtype=torch.float32)

        enc = self.tokenizer(
            text,
            padding="max_length",
            truncation=True,
            max_length=self.max_len,
            return_tensors="pt"
        )

        item = {
            "input_ids": enc["input_ids"].squeeze(0),
            "attention_mask": enc["attention_mask"].squeeze(0),
            "meta": meta
        }

        if self.has_labels:
            item["labels"] = torch.tensor(self.labels[idx], dtype=torch.long)

        return item

# Dataset labeled untuk UST
train_labeled_ds_ust = ReviewDataset(train_labeled, train_meta, tokenizer, MAX_LEN, has_labels=True)
val_ds_ust           = ReviewDataset(val_labeled,   val_meta,   tokenizer, MAX_LEN, has_labels=True)
test_ds_ust          = ReviewDataset(test_labeled,  test_meta,  tokenizer, MAX_LEN, has_labels=True)

BATCH_LABELED = 8
BATCH_UNLABELED = 16

train_labeled_loader_ust = DataLoader(train_labeled_ds_ust, batch_size=BATCH_LABELED, shuffle=True)
val_loader_ust           = DataLoader(val_ds_ust,          batch_size=64, shuffle=False)
test_loader_ust          = DataLoader(test_ds_ust,         batch_size=64, shuffle=False)

print("Batches train labeled:", len(train_labeled_loader_ust))
```

## Model fusion

```python
class FusionClassifier(nn.Module):
    def __init__(self, bert_model, meta_dim, hidden_size=256, num_labels=2):
        super().__init__()
        self.bert = bert_model
        bert_hidden = bert_model.config.hidden_size

        self.classifier = nn.Sequential(
            nn.Linear(bert_hidden + meta_dim, hidden_size),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(hidden_size, num_labels)
        )

    def forward(self, input_ids, attention_mask, meta):
        outputs = self.bert(
            input_ids=input_ids,
            attention_mask=attention_mask
        )
        cls_emb = outputs.last_hidden_state[:, 0, :]  # token [CLS]
        x = torch.cat([cls_emb, meta], dim=1)
        logits = self.classifier(x)
        return logits

ust_model = FusionClassifier(bert_base, meta_dim=meta_dim).to(device)

from torch.optim import AdamW
optimizer_ust = AdamW(ust_model.parameters(), lr=2e-5)
ce_loss = nn.CrossEntropyLoss()
```

## Fungsi evaluasi

```python
def evaluate(model, loader, name="MODEL"):
    model.eval()
    all_preds, all_probs, all_labels = [], [], []

    with torch.no_grad():
        for batch in loader:
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            meta = batch["meta"].to(device)
            labels = batch["labels"].to(device)

            logits = model(input_ids, attention_mask, meta)
            probs = F.softmax(logits, dim=1)[:, 1]
            preds = torch.argmax(logits, dim=1)

            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())

    print("\n=== Evaluation:", name, "===")
    print(classification_report(all_labels, all_preds, digits=4))
    try:
        print("ROC-AUC:", roc_auc_score(all_labels, all_probs))
    except:
        print("ROC-AUC tidak dapat dihitung (label hanya 1 kelas).")
```

## Loop UST

```python
NUM_EPOCHS = 5
CONF_THRESH = 0.9

for epoch in range(NUM_EPOCHS):
    print(f"\n========== UST Epoch {epoch+1}/{NUM_EPOCHS} ==========")

    # 1. Sampling 500 unlabeled
    n_unlab = min(500, len(unlabeled_df))
    unlabeled_small = unlabeled_df.sample(
        n=n_unlab,
        random_state=42 + epoch,
        replace=False
    ).reset_index(drop=True)

    # 2. Dataset untuk unlabeled
    unlabeled_meta_small = make_meta_matrix(unlabeled_small)
    unlabeled_ds_ust = ReviewDataset(
        unlabeled_small,
        unlabeled_meta_small,
        tokenizer,
        MAX_LEN,
        has_labels=False
    )
    unlabeled_loader_ust = DataLoader(
        unlabeled_ds_ust,
        batch_size=BATCH_UNLABELED,
        shuffle=False
    )

    # 3. Prediksi pseudo-label di unlabeled_small
    ust_model.eval()
    all_probs, all_preds = [], []

    with torch.no_grad():
        for batch in unlabeled_loader_ust:
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            meta = batch["meta"].to(device)

            logits = ust_model(input_ids, attention_mask, meta)
            probs = F.softmax(logits, dim=1)
            max_probs, preds = torch.max(probs, dim=1)

            all_probs.extend(max_probs.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())

    all_probs = np.array(all_probs)
    all_preds = np.array(all_preds)

    # 4. Filter pseudo-label dengan confidence tinggi
    mask_conf = all_probs >= CONF_THRESH
    n_conf = mask_conf.sum()

    print(f"Unlabeled sampled: {n_unlab} | Confident pseudo-labels: {n_conf} (th={CONF_THRESH})")

    if n_conf == 0:
        # Tidak ada pseudo-label yang confident, pakai train_labeled saja
        train_combined_df = train_labeled.copy()
        train_combined_meta = train_meta
    else:
        pseudo_df = unlabeled_small.loc[mask_conf].copy()
        pseudo_df["label"] = all_preds[mask_conf]

        train_combined_df = pd.concat([train_labeled, pseudo_df], ignore_index=True)
        train_combined_meta = make_meta_matrix(train_combined_df)

    print("Train combined size:", len(train_combined_df))

    # 5. DataLoader gabungan
    train_combined_ds = ReviewDataset(
        train_combined_df,
        train_combined_meta,
        tokenizer,
        MAX_LEN,
        has_labels=True
    )
    train_combined_loader = DataLoader(
        train_combined_ds,
        batch_size=BATCH_LABELED,
        shuffle=True
    )

    # 6. Training 1 epoch di data gabungan
    ust_model.train()
    total_loss = 0.0

    for batch in train_combined_loader:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        meta = batch["meta"].to(device)
        labels = batch["labels"].to(device)

        optimizer_ust.zero_grad()

        logits = ust_model(input_ids, attention_mask, meta)
        loss = ce_loss(logits, labels)

        loss.backward()
        optimizer_ust.step()

        total_loss += loss.item()

    avg_loss = total_loss / len(train_combined_loader)
    print(f"Train loss (combined): {avg_loss:.4f}")

    # 7. Evaluasi di validation set
    evaluate(ust_model, val_loader_ust, name=f"UST Epoch {epoch+1} (Val)")
```

## Evaluasi

```python
print("\n===== Evaluasi AKHIR UST di Test Set =====")
evaluate(ust_model, test_loader_ust, name="UST Final (Test)")
```

# Unlabeled data prediction

## DataLoader

```python
# Siapkan metadata untuk seluruh unlabeled
unlabeled_meta_full = make_meta_matrix(unlabeled_df)

unlabeled_ds_full = ReviewDataset(
    unlabeled_df,
    unlabeled_meta_full,
    tokenizer,
    MAX_LEN,
    has_labels=False
)

unlabeled_loader_full = DataLoader(
    unlabeled_ds_full,
    batch_size=32,
    shuffle=False
)
```

## Predict

```python
ust_model.eval()

all_preds = []
all_probs = []

with torch.no_grad():
    for batch in unlabeled_loader_full:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        meta = batch["meta"].to(device)

        logits = ust_model(input_ids, attention_mask, meta)
        probs = F.softmax(logits, dim=1)

        pred = torch.argmax(probs, dim=1)
        p1 = probs[:, 1]    # probabilitas kelas 1 (manipulatif)

        all_preds.extend(pred.cpu().numpy())
        all_probs.extend(p1.cpu().numpy())
```

## Add pseudo-label to dataframe unlabeled

```python
unlabeled_df_pred = unlabeled_df.copy()
unlabeled_df_pred["pseudo_label"] = all_preds
unlabeled_df_pred["pseudo_prob"] = all_probs
```
